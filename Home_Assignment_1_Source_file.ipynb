**Task: Tensor Reshaping & Operations**
1.	Create a random tensor of shape (4, 6).
2.	Find its rank and shape using TensorFlow functions.
3.	Reshape it into (2, 3, 4) and transpose it to (3, 2, 4).
4.	Broadcast a smaller tensor (1, 4) to match the larger tensor and add them.
5.	Explain how broadcasting works in TensorFlow.


Expected Output:
a)	Print rank and shape of the tensor before and after reshaping/transposing.


```python
import tensorflow as tf

# Step 1: Create a random tensor of shape (4, 6)
tensor = tf.random.uniform(shape=(4, 6))

# Step 2: Find rank and shape using TensorFlow functions
rank = tf.rank(tensor).numpy()
shape = tensor.shape
print(f"Original Tensor Rank: {rank}, Shape: {shape}")

# Step 3: Reshape it into (2, 3, 4) and transpose to (3, 2, 4)
reshaped_tensor = tf.reshape(tensor, (2, 3, 4))
transposed_tensor = tf.transpose(reshaped_tensor, perm=[1, 0, 2])
print(f"Reshaped Tensor Shape: {reshaped_tensor.shape}")
print(f"Transposed Tensor Shape: {transposed_tensor.shape}")

# Step 4: Broadcasting a smaller tensor (1, 4) and adding
small_tensor = tf.random.uniform(shape=(1, 4))
broadcasted_tensor = small_tensor + tf.zeros_like(reshaped_tensor)
result = reshaped_tensor + broadcasted_tensor
print(f"Broadcasted Tensor Shape: {broadcasted_tensor.shape}")
print(f"Resultant Tensor Shape: {result.shape}")

# Step 5: Explanation of Broadcasting in TensorFlow
broadcasting_explanation = """
Broadcasting allows TensorFlow to perform element-wise operations on tensors with different shapes.
When a smaller tensor is added to a larger one, TensorFlow automatically expands the smaller tensor's dimensions
if they are compatible. In this case, (1,4) was broadcasted to match (2,3,4) by expanding the first two dimensions.
"""
print(broadcasting_explanation)
```

2.

  **Loss Functions & Hyperparameter Tuning**


Task: Implement and Compare Loss Functions
1.	Define true values (y_true) and model predictions (y_pred).
2.	Compute Mean Squared Error (MSE) and Categorical Cross-Entropy (CCE) losses.
3.	Modify predictions slightly and check how loss values change.
4.	Plot loss function values using Matplotlib.

**Expected Output**:

•	Loss values printed for different predictions.

•	Bar chart comparing MSE and Cross-Entropy Loss.


```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Step 1: Define true values (y_true) and model predictions (y_pred)
y_true = tf.constant([0, 1, 1, 0], dtype=tf.float32)
y_pred1 = tf.constant([0.1, 0.9, 0.8, 0.2], dtype=tf.float32)
y_pred2 = tf.constant([0.2, 0.8, 0.7, 0.3], dtype=tf.float32)

# Step 2: Compute Mean Squared Error (MSE) and Categorical Cross-Entropy (CCE) losses
mse_loss1 = tf.keras.losses.MeanSquaredError()(y_true, y_pred1).numpy()
mse_loss2 = tf.keras.losses.MeanSquaredError()(y_true, y_pred2).numpy()
cce_loss1 = tf.keras.losses.CategoricalCrossentropy()(tf.expand_dims(y_true, axis=0), tf.expand_dims(y_pred1, axis=0)).numpy()
cce_loss2 = tf.keras.losses.CategoricalCrossentropy()(tf.expand_dims(y_true, axis=0), tf.expand_dims(y_pred2, axis=0)).numpy()

# Step 3: Print loss values
print(f"MSE Loss (Prediction 1): {mse_loss1}")
print(f"MSE Loss (Prediction 2): {mse_loss2}")
print(f"CCE Loss (Prediction 1): {cce_loss1}")
print(f"CCE Loss (Prediction 2): {cce_loss2}")

# Step 4: Plot loss function values using Matplotlib
loss_types = ['MSE - Pred1', 'MSE - Pred2', 'CCE - Pred1', 'CCE - Pred2']
loss_values = [mse_loss1, mse_loss2, cce_loss1, cce_loss2]

plt.bar(loss_types, loss_values, color=['blue', 'blue', 'red', 'red'])
plt.ylabel('Loss Value')
plt.title('Comparison of MSE and CCE Loss')
plt.show()
```

3.

**Train a Model with Different Optimizers**

**Task: Train MNIST Model with Adam & SGD**
1.	Load the MNIST dataset.
2.	Train two models: One with Adam and another with SGD.
3.	Compare training and validation accuracy trends.

**Expected Output:**

•	Accuracy plots comparing Adam vs. SGD performance


```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam, SGD

# Step 1: Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Step 2: Define a function to build the model
def build_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
    return model

# Train model with Adam
adam_model = build_model()
adam_model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
adam_history = adam_model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, verbose=1)

# Train model with SGD
sgd_model = build_model()
sgd_model.compile(optimizer=SGD(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
sgd_history = sgd_model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, verbose=1)

# Step 3: Compare training and validation accuracy trends
plt.plot(adam_history.history['accuracy'], label='Adam Train')
plt.plot(adam_history.history['val_accuracy'], label='Adam Val')
plt.plot(sgd_history.history['accuracy'], label='SGD Train')
plt.plot(sgd_history.history['val_accuracy'], label='SGD Val')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Adam vs. SGD Accuracy Comparison')
plt.show()
```

4.

**Train a Neural Network and Log to TensorBoard
Task Description**
1.	Load the MNIST dataset and preprocess it.
2.	Train a simple neural network model and enable TensorBoard logging.
3.	Launch TensorBoard and analyze loss and accuracy trends.

**Expected Outcome:**

•	The model should train for 5 epochs and store logs in the "logs/fit/" directory.

•	You should be able to visualize training vs. validation accuracy and loss in TensorBoard.


```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.callbacks import TensorBoard
import datetime

# Step 1: Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Step 2: Define a function to build the model
def build_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax')
    ])
    return model

# Create TensorBoard callback
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)

# Train model with Adam
adam_model = build_model()
adam_model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
adam_history = adam_model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, verbose=1, callbacks=[tensorboard_callback])

# Train model with SGD
sgd_model = build_model()
sgd_model.compile(optimizer=SGD(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
sgd_history = sgd_model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=5, verbose=1, callbacks=[tensorboard_callback])

# Step 3: Compare training and validation accuracy trends
plt.plot(adam_history.history['accuracy'], label='Adam Train')
plt.plot(adam_history.history['val_accuracy'], label='Adam Val')
plt.plot(sgd_history.history['accuracy'], label='SGD Train')
plt.plot(sgd_history.history['val_accuracy'], label='SGD Val')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Adam vs. SGD Accuracy Comparison')
plt.show()

# Instructions to launch TensorBoard
print("Run the following command to launch TensorBoard:")
print("tensorboard --logdir=logs/fit")
```
